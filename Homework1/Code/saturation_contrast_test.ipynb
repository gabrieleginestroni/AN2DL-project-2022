```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
temp = Image.open('training_data_final/Species1/00000.jpg').convert('RGB')
image = np.squeeze(np.expand_dims(temp, axis=0))
fig, ax = plt.subplots(1, 1)
ax.imshow(Image.fromarray(image.astype(np.uint8)))
plt.tight_layout()
plt.show()
```

# Saturation

```python
num_images = 4

fig, axes = plt.subplots(1, num_images)
axes[0].imshow(Image.fromarray(image.astype(np.uint8)))

for i in range(1, num_images):
    ax = axes[i]
    ax.imshow(Image.fromarray(np.array(tf.image.adjust_saturation(image, 1 + i))))

plt.tight_layout()
plt.show()
```

# Contrast

```python
num_images = 4

fig, axes = plt.subplots(1, num_images)
axes[0].imshow(Image.fromarray(image.astype(np.uint8)))

for i in range(1, num_images):
    ax = axes[i]
    ax.imshow(Image.fromarray(np.array(tf.image.adjust_contrast(image, 1 + i/4))))

plt.tight_layout()
plt.show()
```
